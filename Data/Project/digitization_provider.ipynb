```python
import pandas as pd
import numpy as np
```

```python
df_semi=pd.read_csv('movie_with_rat_act.csv',index_col=0)

df_provider1 = pd.read_csv('provider_setting2.csv',index_col=0)
df_provider2 = pd.read_csv('provider_setting1.csv',index_col=0)
```

```python
df_semi.head()
```

```python
df_provider1=df_provider1.drop(['guest','cnt','averageG','guest_N','averageG_N','totalsum','cnt_N','averagesum'],axis=1)
```

```python
df_provider2=df_provider2.drop(['guest','cnt','averageG','guest_N','averageG_N','totalsum','cnt_N','averagesum'],axis=1)
```

```python
df_provider2= df_provider2.rename(columns={'Unnamed: 11':'cutting'})
```

```python
df_provider1= df_provider1.rename(columns={'provider_grade' : 'cutting'})
```

```python
df_provider1.
```

```python
df_pmodel1 = df_provider1.dropna()
```

```python
df_pmodel2 = df_provider2.dropna()
```

```python
df_pmodel1.index =list(range(len(df_pmodel1)))
df_pmodel2.index =list(range(len(df_pmodel2)))
df_semi.index=list(range(len(df_semi)))
```

```python
df_semi1 = df_semi.drop(['Unnamed: 0.1','money'],axis=1)
```

```python
df_semi.head()
df_semi1['setting1_power']=''
df_semi1['setting1_grade']=''

df_semi1['setting2_power']=''
df_semi1['setting2_grade']=''
```

```python
for semidx in range(len(df_semi1)):
    plist = df_semi1['provider'][semidx].split(',')
    sum=0
    sum2=0
    if len(plist) == 1:
        if len(df_pmodel1[df_pmodel1['pname']==plist[0]]['Tot_Ave_cnt_sum']) == 0: # semi에서 찾을 사람이 없음
            df_semi1.loc[semidx,['setting1_grade']] = 'Na'
            df_semi1.loc[semidx,['setting1_power']] = 'Na'
            continue
            
        df_semi1.loc[semidx,['setting1_power']] = df_pmodel1[df_pmodel1['pname'] == plist[0]]['Tot_Ave_cnt_sum'].values[0]
        df_semi1.loc[semidx,['setting1_grade']] = df_pmodel1[df_pmodel1['pname'] == plist[0]]['cutting'].values[0]

    else:
        cnt = len(plist)
        for provider in plist:
            if len(df_pmodel1[df_pmodel1['pname']==provider]) == 0: # semi에서 찾을 사람이 없음
                cnt = cnt - 1
                continue
            else:
                sum= sum + df_pmodel1[df_pmodel1['pname'] == provider]['Tot_Ave_cnt_sum'].values[0]
                sum2 = sum2 + df_pmodel1[df_pmodel1['pname'] == provider]['cutting'].values[0]
        if sum == 0:
            continue
        df_semi1.loc[semidx,['setting1_power']] = sum / cnt
        df_semi1.loc[semidx,['setting1_grade']] = sum2 / cnt
    print("=====================================")
    print(df_semi1['mname'][semidx])
    print("=====================================")
print("complete")
```
