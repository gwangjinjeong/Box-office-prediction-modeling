```python
import time, re, requests
from bs4 import BeautifulSoup
import pandas as pd
import csv
import datetime
import calendar

menuarr = ['날짜','영화명','매출액','매출액점유율','매출액증감','누적매출액','관객수','관객수증감','누적관객수','스크린수','상영횟수']

df = pd.DataFrame(columns=menuarr)
now = datetime.datetime.now()
base_url="http://www.kobis.or.kr/kobis/business/stat/boxs/findDailyBoxOfficeList.do?loadEnd=0&searchType=search&sSearchFrom="
year = 2018
for month in range(1,13):
    if now.month < month:
        break
    for day in range(1,calendar.monthrange(year,month)[1]+1):
        if (now.year == year) & (now.month == month) & (now.day == day):
            break
        url = base_url+str(year)+"-" + str(month).zfill(2)+"-" + str(day).zfill(2) + "&sSearchTo="+str(year)+"-"+str(month).zfill(2)+"-" + str(day).zfill(2)    
        source_code = requests.get(url)
        plain_text = source_code.text
        soup = BeautifulSoup(plain_text, 'html.parser' )
        #초기화###################################
        rowdata = []
        titledata=[]
        figuredata=[]
        list_of_lists =[]
        #일자#####################################
        date = soup.find('div', attrs={'class': 'board_tit'})
        print(date.text) # 일자
        #row######################################
        content = soup.find('div',attrs={'class':'listArea'})
        content2 = content.findAll('th', attrs={'scope': 'col'}) 
        for i in content2:
            i1 = i.text.replace("\t", "").replace("오름차순","").replace("내림차순","")
            rowdata.append(i1.strip())
        #영화제목#################################
        tag_body = soup.find('tbody', attrs={'id': 'tbody_0'})
        tags_title = tag_body.findAll('td', attrs={'class': 'title'}) # 타이틀 전체
        for i in tags_title:
            i1 = re.sub('[동일,New]','',i.text)
            i2 = re.sub('([0-9]+)상승','',i1)
            i3 = re.sub('([0-9]+)하락','',i2)
            titledata.append(i3.strip())
        #영화매출액###############################
        tags_rel = tag_body.findAll('td', attrs={'class': 'figureR'})
        for i in tags_rel:
            i1 = i.text.replace('\t','')
            figuredata.append(i1.strip())
        #####################################################
        for i in range(10):
            list_of_lists.append([date.text.strip(),titledata[i]])
            list_of_lists[i].extend(figuredata[i*9:i*9+9])
        df2 = pd.DataFrame(data=list_of_lists,columns=menuarr)
        df = df.append(df2)
```

```python
df.to_excel('test1.xlsx',sheet_name='Sheet1')
```
