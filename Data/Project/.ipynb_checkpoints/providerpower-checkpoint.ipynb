```python
import numpy as np
import pandas as pd
```

```python
df_semi=pd.read_csv('total boxoffice12_2_18_7c.csv',skiprows=2)
```

```python
df_provider=pd.DataFrame({"pname": "a", "guest": 0,"cnt":0},columns=['pname','guest','cnt'],index=[1])#['a','b','c'], index=[0,2,4]
```

```python
df_semi2= df_semi.drop(['profit','movieCd','profitPercent','profitAcc','nation','nations','maker','actor','director','rating ',' genre','audiCnt','scrnCnt','showCnt'],axis=1)
```

```python
df_semi2 = df_semi2.dropna()
```

```python
df_semi2.index =list(range(len(df_semi2)))
```

```python
idx=0
for pvr in df_semi2['provier']: #pvr은 semi에서 provider의 각 항의 목록을 가지고옴
    if idx == 1916:
        break
    print(df_semi2['movieNm'][idx])
    pvrlist = pvr.split(',')     # 그 각 항을 ,으로 split
    for part_of_pvr in pvrlist: # 쪼개진 항에서 하나씩 가지고 옴
        sum=0
        cnt=0
        if part_of_pvr in list(df_provider['pname']):
            pname_idx = df_provider[df_provider['pname'] == part_of_pvr].index.tolist()[0]
            sum= df_provider['guest'][pname_idx] + df_semi2['audiAcc'][idx]
            df_provider.loc[pname_idx,['guest']] = sum
            df_provider.loc[pname_idx,['cnt']] = df_provider['cnt'][pname_idx] + 1
        else:
            new_idx = len(df_provider)+1
            df_provider.loc[new_idx,['pname']] = part_of_pvr
            df_provider.loc[new_idx,['guest']] = df_semi2['audiAcc'][idx]
            cnt=cnt+1
            df_provider.loc[new_idx,['cnt']] = cnt
        idx=idx+1
print("==================================complete==================================")
#df_semi2[df_semi2['mname'] == word].index.tolist()[0]
```
