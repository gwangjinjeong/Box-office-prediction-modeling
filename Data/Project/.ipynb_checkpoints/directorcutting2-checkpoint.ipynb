```python
for act_idx in range(len(df_semi1)):
    
    dnamelist = df_semi1['director'][act_idx].split(',')
    mname = df_semi1['mname'][act_idx]
    sum=0
    
    for dname in dnamelist: # 값이 없으면 실행조차 되지 않았을 것.
        cnt=len(dnamelist)
        if len(df_dmodel[df_dmodel['peopleNm']==dname]['Tot_median']) == 0: # semi에서 찾을 사람이 없음
            cnt=cnt - 1
            continue;
        elif len(df_dmodel[df_dmodel['peopleNm']==dname]['Tot_median']) == 1: # semi에서 찾은사람이 유일함
            Tot_data = df_dmodel[df_dmodel['peopleNm']==dname]['Tot_median'].values[0]
            sum = Tot_data + sum
            df_semi1.loc[act_idx,['power']] = sum/cnt
        else:                                                                  # semi에서 동명이인이 있다!
            for i in df_dmodel[df_dmodel['peopleNm']==df_semi1['director'][act_idx]].index.tolist():
                if mname in df_dmodel['filmoNames'][i].split('|'):
                    print('if문 돌아가는중')
                    df_semi1.loc[act_idx,['power']] = float(df_dmodel['Tot_median'][i])
                print("=====================================")
                print(df_dmodel['peopleNm'][i])
                print(df_dmodel['Tot_median'][i])
                print("=====================================")
    
print('complete')     
#첫번쨰 시행
# 두번째 시행 null 중단?
# 첫번쨰 null
# 두번째 값이 있으면?
```
