```python
import pandas as pd
import numpy as np
```

```python
df_actor=pd.read_csv('actor.csv',index_col=0)
df_semi =pd.read_csv('movie_with_rat_act.csv',index_col=0)
```

```python
df_semi2=df_semi.drop('Unnamed: 0.1',axis=1)
```

```python
df_actor2=df_actor.dropna()
```

```python
df_actor2['guest']= "Na"
df_actor2['movie_num'] ="Na"
for act_idx in range(len(df_actor2['filmoNames'])):
    sum=0
    cnt=0
    search=df_actor2['filmoNames'][act_idx].split('|') # 여기에는 | 으로
    print("----------------------wordsearch---------------------------")
    print(str(df_actor2['peopleNm'][act_idx])+"\t\tsearch data \t"+str(search))
    for word in search:
        if len(df_semi2[df_semi2['mname'] == word]) != 0:
            print("--------------------namesearching-----------------------------")
            semilist = df_semi2[df_semi2['mname'] == word]['actor'].values[0].split('|')
            print("--------------------namesearching-----------------------------")
            print("namelist\t" +str(semilist))
        else:
            continue
        act_name = df_actor2['peopleNm'][act_idx]
        if act_name in semilist:
            guest_idx=df_semi2[df_semi2['mname'] == word].index.tolist()[0]
            cnt = cnt+1
            sum = sum+df_semi2['guest'][guest_idx]
        df_actor2.set_value(act_idx,'guest',sum)
        df_actor2.set_value(act_idx,'movie_num',cnt)
```

```python
df_actor2.to_excel('actor_extension.xlsx',sheet_name='Sheet1')
```
