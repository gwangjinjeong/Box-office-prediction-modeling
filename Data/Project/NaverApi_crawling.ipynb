```python
client_id = "D4rJqpB5sQrGOQZ098Xk"
client_secret = "PhSvV07Pho"
### 다른거 입력하게 test['영화명']
# #title = []
director = []
userrating=[]
actor=[]

keyword=movie['영화명']
# for name in move['영화명']:
#     encText = urllib.parse.quote(name)
#     url = ''

cnt = 0
for name_cnt in range(len(movie['영화명'])):
    encText = urllib.parse.quote(keyword[name_cnt])
    url = "https://openapi.naver.com/v1/search/movie.xml?query=" + encText # xml 결과

    request = urllib.request.Request(url)
    request.add_header("X-Naver-Client-Id",client_id)
    request.add_header("X-Naver-Client-Secret",client_secret)
    response = urllib.request.urlopen(request)
    rescode = response.getcode()
    #print("----"+"mid"+"----")
    if(rescode==200):
        response_body = response.read().decode('utf-8')
        soup2 = BeautifulSoup(response_body,"html.parser")
        sleep(0.05)
        #print(soup2.prettify())
        
        for dir_cnt in soup2.find_all('director'):
            director.append(dir_cnt.text.replace("|","").rstrip(','))
            
        for usr_rat in soup2.findAll('userrating'): 
            userrating.append(usr_rat.text) 
        for act_name in soup2.findAll('actor'):
            actor.append(act_name.text)
        #for i in soup2.find_all('director'):
        #    print(director)

        for dir_cor in range(len(director)):
            if(director[dir_cor] == movie[keyword==keyword[name_cnt]]['감독'].values[0]):
                movie.set_value(name_cnt,'평점',userrating[dir_cor])
                movie.set_value(name_cnt,'배우',actor[dir_cor])
    print(name_cnt)
print("-----------complete------------")
```
