```python
import sys
import urllib
from urllib.request import urlopen
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup

rawurl = 'http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieList.xml?key=74f9b12628513b6fcc6cba6562647217&openStartDt=2011&openEndDt=2018&itemPerPage=885&curPage='
print("---------------receive rawurl---------------")
df_movie=[]
df_movie = pd.DataFrame()
df_movie = df_movie.append({"movieCd":"","movieNm":"","openDt":"","repNationNm":"","repGenreNm":"","peopleNm":""}, ignore_index=True)

print("---------------making df complete---------------")
cnt=0
for page in range(1,11):
    print("---------------{} page range start---------------".format(page))
    movieurl = rawurl + str(page)
    movieres = urlopen(movieurl)
    soup = BeautifulSoup(movieres, "html.parser")
    
    for i in range(0,885):
        df_movie.set_value(cnt, "movieCd", soup.find_all("moviecd")[i].text)
        df_movie.set_value(cnt, "movieNm", soup.find_all("movienm")[i].text)
        df_movie.set_value(cnt, "openDt", soup.find_all("opendt")[i].text)
        df_movie.set_value(cnt, "repNationNm", soup.find_all("repnationnm")[i].text)
        df_movie.set_value(cnt, "repGenreNm", soup.find_all("repgenrenm")[i].text)
        if soup.find_all("directors")[i].find('peoplenm') == None:
            df_movie.set_value(cnt, "peopleNm","")
        else :
            df_movie.set_value(cnt, "peopleNm", soup.find_all("directors")[i].find('peoplenm').text)
        cnt=cnt+1
            
    print("---------------{} page range end---------------".format(page))
print("--------------Complete_^^--------------")
```

```python
df_movie.to_excel('movie.xlsx',sheet_name='Sheet1')
```
