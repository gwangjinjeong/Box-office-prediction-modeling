```
!pip install selenium
!apt-get update
!apt-get install -y chromium-chromedriver
# 이거 안해주면 에러남
import pandas as pd
import time
from bs4 import BeautifulSoup
from selenium import webdriver
import re
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException
from IPython.display import Image

# urllib3.exceptions.maxretryerror httpsconnectionpool
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('/usr/lib/chromium-browser/chromedriver',options=options)
```

```
%%time
class GetKOFICData():
  def __init__(self,driver,url,fname):
    self.driver = driver
    self.fname  = fname
    self.url    = url
    self.cnt = 1
    self.last = False 
    self.total = pd.DataFrame()
    self.rest = 0
    self.setup()

  def setup(self):
    self.driver.get(url)
    html = self.driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    self.page = self.preprocessing(pd.read_html(html)[0],
                                   soup.findAll('span', attrs={'class': 'ellip'}))
    self.total = pd.concat([self.total, self.page], axis = 0)

  def getData(self, button):
    print(self.cnt)
    temp = self.page.copy()
    while pd.DataFrame.equals(temp, self.page):
      try:
        self.driver.find_element_by_css_selector(button).click()
        html = self.driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        temp = self.preprocessing(pd.read_html(html)[0],
                                  soup.findAll('span', attrs={'class': 'ellip'}))
      except StaleElementReferenceException:
        time.sleep(4)
        self.rest = self.rest + 1
        self.getData(button)
        pass
      except NoSuchElementException:
        self.page = pd.DataFrame()
        self.total.to_csv(str(self.fname), index = True)
        self.last = not self.last
        break
    self.cnt = self.cnt + 1
    self.page = temp
    self.total = pd.concat([self.total, self.page], axis = 0)

  def preprocessing(self,page,filmo):
    rowNum = int(len(filmo) / 3)
    for i, item in zip(range(rowNum),range(2,rowNum*3,3)):
      page.loc[i,'필모그래피'] = filmo[item].get('title')
    return page

  def run(self):
    while not self.last:
      for pageNum in range(2,11):
        self.getData(f'#pagingForm > div > ul > li:nth-child({pageNum}) > a')
      self.getData('#pagingForm > div > a.btn.next')

  def main(self):
    self.run() 

if __name__=="__main__":
  url = 'http://www.kobis.or.kr/kobis/business/mast/peop/searchPeopleList.do'
  fname = 'Film_person.csv'
  a = GetKOFICData(driver,url,fname)
  a.main()
```
