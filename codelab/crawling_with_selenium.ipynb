```
from selenium import webdriver
import pandas as pd

browser = webdriver.PhantomJS()
browser.implicitly_wait(3)
url = "http://www.kobis.or.kr/kobis/business/mast/peop/searchPeopleList.do"
browser.get(url)
print('페이지접근')
e = browser.find_element_by_name("sPeopleNm")
e.clear()
e.send_keys('이병헌')
print('배우명 입력')
form = browser.find_element_by_css_selector("button.btn_type03[type=button]")
form.submit()
print('브라우저 넘겨주기')
actFeat = []
names = browser.find_elements_by_css_selector("#content > table > tbody > tr > td")
print(names)
idx=0
for name in names:
  if(name.text == '배우'):
    actFeat.append('//'.join(map(lambda x: x.text, names[idx-2:idx+4])))
  idx= idx+1
print(actFeat)
```
