```
!pip install selenium
!apt-get update
!apt-get install -y chromium-chromedriver
# 이거 안해주면 에러남
import pandas as pd
import time
from bs4 import BeautifulSoup
from selenium import webdriver
import re
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException
from IPython.display import Image

# urllib3.exceptions.maxretryerror httpsconnectionpool
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('/usr/lib/chromium-browser/chromedriver',options=options)
```

```
%%time
class GetKOFICCompanyData():
  def __init__(self,driver,url,fname):
    self.driver = driver
    self.fname  = fname
    self.url    = url
    self.cnt = 1
    self.last = False 
    self.total = pd.DataFrame()
    self.rest = 0
    self.tempDic = dict()
    self.tempDF = pd.DataFrame()
    self.contents = ['코드','A.K.A','분류','대표자','영업상태','주소','대표연락처','팩스']
    self.setup()

  def setup(self):
    self.driver.get(url)
    html = self.driver.page_source
    self.getClickedData(html)

  def getClickedData(self,html):
    for itemNum in range(1,11):
      try:
        driver.find_element_by_xpath(f'//*[@id="content"]/div[4]/div[2]/table/tbody/tr[1]/td[{itemNum}]/span/a').click()
        html = self.driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        preprocessed = soup.find('dl', attrs={'class': 'ovf cont'})
        items = re.sub(re.compile(r'\s+'), '/', preprocessed.text).split('/')
        for col in range(len(self.contents)-1):
          self.tempDic[self.contents[col]] = ''.join(items[items.index(self.contents[col])+1:items.index(self.contents[col+1])])
        self.tempDF = pd.DataFrame.from_dict([self.tempDic])
        self.total = pd.concat([self.total, self.tempDF], axis = 0)
        driver.find_element_by_class_name('close').click()
      except StaleElementReferenceException:
        time.sleep(4)
        continue
      except NoSuchElementException:
        time.sleep(2)
        continue
  def getData(self, button):
    temp = self.page.copy()
    while pd.DataFrame.equals(temp, self.page):
      try:
        self.driver.find_element_by_css_selector(button).click()
        html = self.driver.page_source
        temp = pd.read_html(html)[0]
      except StaleElementReferenceException:
        time.sleep(4)
        self.rest = self.rest + 1
        pass
      except NoSuchElementException:
        return 0
    try:
      for itemNum in range(1,11):
        driver.find_element_by_xpath(f'//*[@id="content"]/div[4]/div[2]/table/tbody/tr[1]/td[{itemNum}]/span/a').click()
        html = self.driver.page_source
        self.getClickedData(html)
    except NoSuchElementException:
      self.total.to_csv(str(self.fname), index = True)
      self.last = not self.last
    self.cnt = self.cnt + 1
    self.page = temp

  def run(self):
    while not self.last:
      for pageNum in range(2,11):
        self.getData(f'#pagingForm > div > ul > li:nth-child({pageNum}) > a')
      self.getData('#pagingForm > div > a.btn.next')

  def main(self):
    self.run() 

if __name__=="__main__":
  url = 'http://www.kobis.or.kr/kobis/business/mast/comp/searchCompanyList.do'
  fname = 'Film_company.csv'
  a = GetKOFICCompanyData(driver,url,fname)
  a.main()
```
